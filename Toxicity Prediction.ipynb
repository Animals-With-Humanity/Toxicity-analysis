```python
from tdc.single_pred import Tox
import pandas as pd
data=Tox(name = 'herg_central', label_name = 'hERG_at_10uM' )
split = data.get_split()
```

```python
from pysmiles import read_smiles
import networkx as nx
data=pd.DataFrame(split['train'])
formula=data['Drug'][1]
len(read_smiles(formula))
for i in range(1,10):
    formula=data['Drug'][i]
    print(len(read_smiles(formula)))
```

```python
G=read_smiles(data['Drug'].values[3])
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(50,100))
for i in range(1,13):
    sp= plt.subplot(12,4, i)
    G=read_smiles(data['Drug'].values[i])
    nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```
